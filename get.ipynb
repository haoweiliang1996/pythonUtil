```python
class haowei_tree:
    def __init__(self, filename):
        def __count(s):
            k = 0
            for c in s:
                if c == '\t':
                    k += 1
                else:
                    return k

        with open(filename, 'r') as f:
            self.f_lines = f.readlines()
            #print(self.f_lines)
            self.n = len(self.f_lines)
            self.pre = [-1 for i in range(self.n)]
            self.level = [[] for i in range(self.n)]
            #print(len(self.level))
            for i in range(len(self.f_lines)):
                line = self.f_lines[i]
                self.level[__count(line)] += [i]

        def __tree(father=-1, son=-1):
            if self.pre[son] != -1:
                return
            self.pre[son] = father
            for i in range(son + 1, self.n):
                if __count(self.f_lines[i]) <= __count(self.f_lines[son]):
                    return
                else:
                    if self.pre[i] == -1:
                        __tree(son, i)

        for i in range(self.n):
            __tree(-1, i)

    def test(self):
        for i in filter(lambda x: len(x) > 0, self.level):
            print(i,end ="")
        s = 5
        while s != -1:
            print(str(s) + '\t' +self.f_lines[s])
            s = self.pre[s]


tree = haowei_tree('DEBnd.txt')
result_url =[tree.f_lines[i].strip() for i in  tree.level[3]]
def get_parrent_and_itself(s):
    temp = []
    if s != -1:
        s = tree.pre[s]
    while s!=-1:
        temp.append(s)
        s = tree.pre[s]
    for i in temp:
        print(tree.f_lines[i])
    temp.reverse()
    return temp
for i in get_parrent_and_itself(3):
    print(tree.f_lines[i])
```
